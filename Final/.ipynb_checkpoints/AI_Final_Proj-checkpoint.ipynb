#### Data scrape wikipedia page to get list of postal codes in Toronto.

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
webpage = requests.get(url)

# Get the page info and find table in the page
soup = BeautifulSoup(webpage.text, 'lxml')
soup
table1 = soup.find('table')
table1
```

```python
# Obtain the postal code
pCodes = []
for i in table1.find_all('p'):
    substr = "Not assigned"
    contents = i.get_text()
    contents2 = i.get_text()
    contents2[0:3]
    print(contents2)
    if substr in contents:
        pass
    else:
        pCodes.append(contents)
```

```python
# Create dataframe
mydata = pd.DataFrame(pCodes, columns=['Postal Codes'])
```

```python
# Retrieve the postal codes of the locations in Toronto
for j in table1.find_all('tr')[1:]:
    row_data = j.find_all('<b>')
    row = [i.text for i in row_data]
    length = len(mydata)
    mydata = mydata.append(row)
```

```python
#Send to csv
mydata.to_csv('TorontoPostalCodes.csv')
```

#### Reading the csv files to get the data. Display the columns to see joinable data easier, and join the datasets.

```python
import pandas as pd
dataCrime = pd.read_csv ('Neighbourhood_Crime_Rates_(Boundary_File)_.csv')
dataNeighProfiles = pd.read_csv ('neighbourhood-profiles-2016-csv.csv')
dataNeigh = pd.read_csv ('Neighbourhoods-4326.csv')

dataCrime.columns
```

```python
dataNeighProfiles.columns
```

```python
dataNeigh.columns
```

```python
crimeAndNeigh = dataCrime.merge(dataNeigh, how='inner', left_on='Neighbourhood', right_on='AREA_DESC')
crimeAndNeigh
```

#### Transpose the neighbourhood profiles dataset as row has the neighbourhoods not the columns like the other datasets, and join to single dataset.

```python

dataNeighProfilesTransposed = dataNeighProfiles.T
crimeProfilesNeigh = dataNeighProfilesTransposed.merge(crimeAndNeigh, how='inner', left_on=0, right_on='Neighbourhood')
crimeProfilesNeigh
```
